## Household Cost Estimator (CLI Version)

This project is a command-line Python application that calculates estimated daily and weekly household utility costs based on user input. It uses predefined rate constants to compute laundry, shower, and dishwasher expenses and includes basic input validation.

```python
# constants
RATES = {
"laundry": 1.5,
"shower": 0.2, 
"dishwasher": 1 
}

# function to calcualte costs
def calculate_daily_cost(laundry_loads, shower_mins, dishwasher_runs):
    laundry_cost = laundry_loads * RATES["laundry"]
    shower_cost = shower_mins * RATES["shower"]
    dishwasher_cost = dishwasher_runs * RATES["dishwasher"]

    total = laundry_cost + shower_cost + dishwasher_cost 

    return {
        "laundry_cost": laundry_cost, 
        "shower_cost": shower_cost,
        "dishwasher_cost": dishwasher_cost,
        "total": total

    }

#input handling 
def get_positive_number(prompt):
    while True:
        try:
            value = float(input(prompt))

            if value < 0:
                print("Value cannot be negative. Try again")
                continue 

            return value 
        except ValueError:
            print("Invalid input. Please enter a number")

# main program
def main():
    print("---Daily Household Cost Estimator---")
    laundry = get_positive_number("Laundry loads: ")
    shower = get_positive_number("Shower minutes: ")
    dishwasher = get_positive_number("Dishwasher runs: ")

    results = calculate_daily_cost(laundry, shower, dishwasher)

    print(f"---Daily Cost Breakdown---")
    print(f"Laundry: ${results['laundry_cost']:.2f}")
    print(f"Shower: ${results['shower_cost']:.2f}")
    print(f"Dishwasher: ${results['dishwasher_cost']:.2f}")
    print(f"Total Daily Cost: ${results['total']:.2f}")
    print(f"Projected Weekly Cost: ${results['total'] * 7:.2f}")


if __name__ == "__main__":
    main()
```
