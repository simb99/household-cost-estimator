## Household Cost Estimator (GUI Version)

This project is a graphical user interface (GUI) version of the Household Cost Estimator built using Tkinter and ttk. The application allows users to input daily household usage for laundry, shower time, and dishwasher runs, and calculates the corresponding daily and weekly costs.

The GUI version enhances the original command-line implementation by providing an interactive interface, structured layout, and real-time result display while maintaining clean separation of business logic and user interface components.

```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox

# ---------------------------
# Constants
# ---------------------------

RATES = {
    "laundry": 1.5,
    "shower": 0.2,
    "dishwasher": 1
}

# ---------------------------
# Business Logic
# ---------------------------

def calculate_daily_cost(laundry_loads, shower_mins, dishwasher_runs):
    laundry_cost = laundry_loads * RATES["laundry"]
    shower_cost = shower_mins * RATES["shower"]
    dishwasher_cost = dishwasher_runs * RATES["dishwasher"]

    total = laundry_cost + shower_cost + dishwasher_cost

    return {
        "laundry_cost": laundry_cost,
        "shower_cost": shower_cost,
        "dishwasher_cost": dishwasher_cost,
        "total": total
    }

# ---------------------------
# Button Function
# ---------------------------

def calculate():
    try:
        laundry = float(laundry_entry.get())
        shower = float(shower_entry.get())
        dishwasher = float(dishwasher_entry.get())

        if laundry < 0 or shower < 0 or dishwasher < 0:
            messagebox.showerror("Error", "Values cannot be negative.")
            return

        results = calculate_daily_cost(laundry, shower, dishwasher)

        result_label.config(
            text=(
                f"Laundry: ${results['laundry_cost']:.2f}\n"
                f"Shower: ${results['shower_cost']:.2f}\n"
                f"Dishwasher: ${results['dishwasher_cost']:.2f}\n"
                f"Total: ${results['total']:.2f}\n"
                f"Weekly Projection: ${(results['total'] * 7):.2f}"
            )
        )

    except ValueError:
        messagebox.showerror("Error", "Please enter valid numbers.")

# ---------------------------
# GUI Setup
# ---------------------------

root = tk.Tk()
root.title("Daily Household Cost Estimator")
root.geometry("420x520")
root.configure(bg="#f4f6f8")

style = ttk.Style()
style.theme_use("clam")

# Title
title = ttk.Label(
    root,
    text="Household Cost Estimator",
    font=("Arial", 16, "bold")
)
title.pack(pady=15)

# Input Frame
input_frame = ttk.Frame(root, padding=20)
input_frame.pack()

# Laundry
ttk.Label(input_frame, text="Laundry Loads").pack(pady=5)
laundry_entry = ttk.Entry(input_frame)
laundry_entry.pack()

# Shower
ttk.Label(input_frame, text="Shower Minutes").pack(pady=5)
shower_entry = ttk.Entry(input_frame)
shower_entry.pack()

# Dishwasher
ttk.Label(input_frame, text="Dishwasher Runs").pack(pady=5)
dishwasher_entry = ttk.Entry(input_frame)
dishwasher_entry.pack()

# Calculate Button
calculate_button = ttk.Button(
    root,
    text="Calculate",
    command=calculate
)
calculate_button.pack(pady=20)

# Result Label
result_label = ttk.Label(
    root,
    text="",
    justify="left",
    font=("Arial", 11)
)
result_label.pack(pady=10)

root.mainloop()
```
